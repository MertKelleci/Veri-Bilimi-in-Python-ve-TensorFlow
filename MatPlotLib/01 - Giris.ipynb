```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [10,20,30,30,30,40,50,60,70,75]
```

```python
weightList = [20,60,80,85,86,87,70,90,95,90]
```

```python
npyAgeList = np.array(ageList)
npyWeightList = np.array(weightList)
```

```python
plt.plot(npyAgeList, npyWeightList, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age/Weight Ratio")
```

```python
numpyList = np.linspace(0,10,20)
numpyList
```

```python
numpyList2 = numpyList ** 2
plt.plot(numpyList,numpyList2, "r")
```

```python
plt.plot(numpyList,numpyList2, "r--")
```

```python
plt.plot(numpyList,numpyList2, "b*-")
```

```python
plt.subplot(1,2,1) # 1 sra 2 kolon 1. grafik
plt.plot(numpyList,numpyList2, "b*-")
plt.subplot(1,2,2) # 1 sra 2 kolon 2. grafik
plt.plot(numpyList2,numpyList, "r--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyList,numpyList2,"g")
figureAxes.set_xlabel("X Ekseni")
figureAxes.set_ylabel("Y Ekseni")
figureAxes.set_title("Grafik Bal")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyList,numpyList2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik")

eksen2.plot(numpyList2,numpyList,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Kk Grafik")
```

```python
(myFigure, myAxes) = plt.subplots()
myAxes.plot(numpyList, numpyList2, "b")
```

```python
(myFigure, myAxes) = plt.subplots(nrows=1, ncols=2)

for axis in myAxes:
    axis.plot(numpyList, numpyList2, "r")
```

```python
newFigure = plt.figure() ##figsize=(tuple)boyut, dpi=(int)znrlk

newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])

newAxis.plot(numpyList,numpyList ** 2, label="numpyList ** 2")
newAxis.plot(numpyList,numpyList ** 3, label="numpyList ** 3")
newAxis.legend(loc=2)
```
